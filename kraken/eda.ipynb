### Historical Data

```python
ohlc = k.get_ohlc_data('BTCUSD', interval=1440, ascending = True)
ohlc[0].head()
```

```python
import pandas as pd

ohlc[0]['20 SMA'] = ohlc[0]['close'].rolling(20).mean()
ohlc_df = ohlc[0] 
ohlc_df.tail()
```

```python
import pandas as pd
import plotly.graph_objects as go


fig = go.Figure(data=[
    go.Candlestick(
        x=ohlc_df.index,
        open=ohlc_df['open'],
        high=ohlc_df['high'],
        low=ohlc_df['low'],
        close=ohlc_df['close']
    ),
    go.Scatter(
        x=ohlc_df.index,
        y=ohlc_df['20 SMA'],
        line=dict(color='purple', width=1),
        name='20 SMA'
    )
])

fig.update_layout(title='BTCUSD Candlestick with 20 SMA')
fig.show()
```

### Get Order Book Data

```python
book = k.get_order_book('AAVEUSD', count = 20)
print(book[0].head())
print(book[1].head())
```

```python
# Merging bid and ask
df = pd.merge(book[0], book[1], left_index=True, right_index=True)
df = df.rename({"price_x":"Bid Price","volume_x":"Bid Amount", "time_x":"Bid Time",
                "price_y":"Ask Price","volume_y":"Ask Amount", "time_y":"Ask Time"}, axis='columns')
df.head()
```
